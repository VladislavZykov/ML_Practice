```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
from skimage import draw
import torch
```

```python
from PIL import Image

image = Image.open('/content/pic.jpg')
```

```python
input_data = np.array(image)
input_data = input_data.transpose((2, 0, 1))
```

```python
image
```

```python
print(input_data.shape)
```

```python
def plot3d(data):
    fig = plt.figure(figsize=(20, 20))
    grid = ImageGrid(fig, 111, nrows_ncols=(1, 3), axes_pad=0.02)
    for ax, im in zip(grid, data):
        ax.imshow(im)
    plt.show()
```

```python
from torch import nn

conv1 = nn.Conv2d(in_channels=3, out_channels=1, kernel_size=(3, 3), stride=1)
input_tensor = torch.from_numpy(input_data).to(torch.float32)
print(input_tensor.size())
output_tensor = conv1(input_tensor)
print(output_tensor.size())
np_out = output_tensor.detach().numpy()
print(np_out.shape)
plot3d(np_out)
```

```python
relu = nn.ReLU()
output_tensor2 = relu(output_tensor)
print(output_tensor2.size())
np_out2 = output_tensor2.detach().numpy()
print(np_out2.shape)
plot3d(np_out2)
```

```python
Sigmoid = nn.Sigmoid()
output_tensor3 = Sigmoid(output_tensor)
print(output_tensor3.size())
np_out3 = output_tensor3.detach().numpy()
print(np_out3.shape)
plot3d(np_out3)
```

```python
Tanh = nn.Tanh()
output_tensor4 = Tanh(output_tensor)
print(output_tensor4.size())
np_out4 = output_tensor4.detach().numpy()
print(np_out4.shape)
plot3d(np_out4)
```

```python
SiLU= nn.SiLU()
output_tensor5 = SiLU(output_tensor)
print(output_tensor5.size())
np_out5 = output_tensor5.detach().numpy()
print(np_out5.shape)
plot3d(np_out5)
```
